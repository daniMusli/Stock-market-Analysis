## Unemployment Rate 

```python
df1 = quandl.get(["FRED/UNRATE"],
    trim_start="2020-1-1", trim_end="2020-11-01", collapse = 'monthly')
df1.columns = ['Unemployment']
#df1.tail()

colss = [col for col in df1.columns]
fig, ax = plt.subplots(1, 1, figsize=(14,6))

for c in colss:
    ax.plot(c, linewidth=2, data=df1)
    ax.set_title('Unemployment Rate')
    plt.xlabel('Year')
    plt.ylabel('Percent')
    handless, labelss = ax.get_legend_handles_labels()
    ax.legend(handless, labelss,loc='upper left', prop={"size":10})    
plt.show()
```
